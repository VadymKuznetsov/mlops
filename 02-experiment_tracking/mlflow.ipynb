```python
'Q2: Download and preprocess data'
'--------------------------------'

import urllib.request
import os
import subprocess

raw_path = 'data/raw'
process_path = 'data/processed'

# Ensure directories exist
os.makedirs(raw_path, exist_ok=True)
os.makedirs(process_path, exist_ok=True)

# Download the raw datasets
for month in ['01', '02', '03']:
    urllib.request.urlretrieve(
        f'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-{month}.parquet',
        f'{raw_path}/green_tripdata_2023-{month}.parquet'
    )

# Process the data using subprocess
subprocess.run([
    'python', 'preprocess_data.py',
    '--raw_data_path', raw_path,
    '--dest_path', process_path
])

# Count files in processed folder
file_count = len([f for f in os.listdir(process_path) if os.path.isfile(os.path.join(process_path, f))])
print(f'Number of files: {file_count}')
```
